# Tokenisasi Kalimat

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')

def tokenisasi_kalimat(text):
    return sent_tokenize(text)

data_sementara = []

for index, row in df.iterrows():
    kalimats = tokenisasi_kalimat(row['abstrak'])
    for kalimat in kalimats:
        data_sementara.append({'index': index, 'abstrak': kalimat})

df = pd.DataFrame(data_sementara)

df
```

# Filtering

```python
df['abstrak'] = df['abstrak'].str.replace('\n', ' ', regex=True)
df['abstrak'] = df['abstrak'].str.replace('[^a-zA-Z\s]', '', regex=True)
df['abstrak'] = df['abstrak'].str.replace('\d+', '', regex=True)
df['abstrak'] = df['abstrak'].replace(r'http\S+', '', regex=True).replace(r'www\S+', '', regex=True)

df
```

# Case Folding

```python
df['abstrak'] = df['abstrak'].str.lower()
df
```

# TF DF IDF

```python
import pandas as pd
import math
from collections import Counter
import re
```

```python
hasil = df.copy()
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.strip()

hasil['abstrak'] = hasil['abstrak'].apply(lambda x: clean_text(' '.join(x)) if isinstance(x, list) else clean_text(x))
tf_documents = [Counter(doc.split()) for doc in hasil['abstrak']]
all_terms = set(term for doc in tf_documents for term in doc)
df_manual = {term: sum(1 for doc in tf_documents if term in doc) for term in all_terms}
N = len(hasil)
idf_manual = {term: math.log(N / (df_manual[term] + 1)) for term in df_manual}

results = []
for term in all_terms:
    tf_values = [doc.get(term, 0) for doc in tf_documents]
    tfidf_values = [tf * idf_manual[term] for tf in tf_values]
    results.append({
        'Kata': term,
        's1': tf_values[0],
        's2': tf_values[1],
        's3': tf_values[2],
        's4': tf_values[3],
        's5': tf_values[4],
        's6': tf_values[5],
        's7': tf_values[6],
        's8': tf_values[7],
        's9': tf_values[8],
        's10': tf_values[9],
        's11': tf_values[10],
        'DF': df_manual[term],
        'Rumus IDF': f"log({N} / ({df_manual[term]} + 1))",
        'Hasil IDF': idf_manual[term],
        'Bobot IDF': tfidf_values
    })

df_hasil_manual = pd.DataFrame(results)
df_hasil_manual.index = df_hasil_manual.index + 1
df_hasil_manual.reset_index(inplace=True)
df_hasil_manual.rename(columns={'index': 'No'}, inplace=True)
df_hasil_manual
```

```python
#sayang dibuang
import pandas as pd
import math
from collections import Counter
import re

hasil = df.copy()

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.strip()

hasil['abstrak'] = hasil['abstrak'].apply(lambda x: clean_text(' '.join(x)) if isinstance(x, list) else clean_text(x))

document = hasil['abstrak'][0].split()

total_terms = len(document)

tf_manual = Counter(document)
for term in tf_manual:
    tf_manual[term] = tf_manual[term] / total_terms

df_manual = {}
N = len(hasil)
for term in tf_manual:
    df_manual[term] = sum(1 for doc in hasil['abstrak'] if term in doc.split())

idf_manual = {}
for term in tf_manual:
    idf_manual[term] = math.log(N / (df_manual[term] + 1))

tfidf_manual = {}
for term in tf_manual:
    tfidf_manual[term] = tf_manual[term] * idf_manual[term]

df_hasil_manual = pd.DataFrame({
    'Kata': list(tf_manual.keys()),
    'TF': list(tf_manual.values()),
    'DF': [df_manual[term] for term in tf_manual],
    'IDF': [idf_manual[term] for term in tf_manual],
    'TF-IDF': [tfidf_manual[term] for term in tf_manual]
})

df_hasil_manual
```
